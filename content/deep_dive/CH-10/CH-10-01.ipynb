   Run your own query and view the result in a chart view

   * To open a new editor, press Ctrl+N.
   * Change the connection context to eltsnap_v2.
   * Type the following query into the query editor:

    ```sql
    SELECT p.project_name
        , COUNT(package_name) AS package_count
    FROM elt.project_package pp
    JOIN elt.project p
        ON p.project_id = pp.project_id
    GROUP BY p.project_name
    ```

   * To execute the query, press F5.
   * After the query results are displayed, click View as Chart, then click the Chart tab.
   * Change Chart Type to Bar. These settings render a bar chart.

**Adding custom insight to the database dashboard**

 ### Step 1 - To open the insight widget configuration, click Create Insight on Chart Viewer

 ### Step 2 - Copy the insight configuration (the JSON data).

 ### Step 3 - Press Ctrl+Comma to open User Settings.

 ### Step 4 - Type dashboard in Search Settings.

 ### Step 5 - Click Edit for dashboard.database.widgets.

 <img src="https://user-images.githubusercontent.com/61198055/75275169-17119b80-57d2-11ea-95ef-161113167c6d.png" width="40%" height="auto">
>

 

 ### Step 6 - Edit the existing dashboard.database.widgets node in your settings file:

 ### Step 7 - Paste the insight configuration JSON into dashboard.database.widgets. Database dashboard settings looks like the following:

  ```json
        "dashboard.database.widgets": [  
         {
        "name":"My-Widget",
        "gridItemConfig":{
            "sizex":2,
            "sizey":1},
            "widget":{
                "insights-widget":
                {
                    "type":
                    {
                        "bar":
                        {
                            "dataDirection":"horizontal",
                            "columnsAsLabels":true,
                            "labelFirstColumn":true,
                            "legendPosition":"bottom",
                            "yAxisLabel":"Package Count",
                            "yAxisMin":-1,
                            "xAxisLabel":"",
                            "encoding":"hex",
                            "imageFormat":"jpeg",
                            "dataType":"number"
                            }
                            },
                            "queryFile":"c:\\Users\\poorvat\\eltQuery2.sql"
                            }
                            }
                        }
     ]
```
